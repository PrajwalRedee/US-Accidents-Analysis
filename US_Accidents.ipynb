## **Data Cleaning**

## **Start Time**

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
plt.title("Percentage of Accidents by hour",size=11)
plt.ylabel('Percentage')
```



*   High Percentage of accidents occur between 6AM to 6PM






```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
plt.title("Percentage of Accidents by Week",size=11)
plt.ylabel('Percentage')
```

```python
df.Start_Time.dt.year.unique()
```

```python
import matplotlib.pyplot as plt
```

Lets check the percentage of accidents for each year.

```python
df_2016 = df[df.Start_Time.dt.year == 2016]
sns.distplot(df_2016.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2016",size=11)
plt.ylabel('Percentage')
```

```python
df_2017 = df[df.Start_Time.dt.year == 2017]
sns.distplot(df_2017.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2017",size=11)
plt.ylabel('Percentage')
```

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
sns.distplot(df_2018.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2018",size=11)
plt.ylabel('Percentage')
```

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2019",size=11)
plt.ylabel('Percentage')
```

```python
df_2020 = df[df.Start_Time.dt.year == 2020]
sns.distplot(df_2020.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2020",size=11)
plt.ylabel('Percentage')
```

```python
df_2021 = df[df.Start_Time.dt.year == 2021]
sns.distplot(df_2021.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
plt.title("Percentage of Accidents in 2021",size=11)
plt.ylabel('Percentage')
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=df.Start_Time.dt.year.value_counts())
plt.title(" Accident cases yearly(2016-2021) occur in US ",size=17,color="grey")
plt.xlabel('Year')
plt.ylabel('No. of accidents')
```

We can see the exponential rise in the number of accidents after 2019

```python
import folium
from folium.plugins import HeatMap
```

```python
sample_df = df.sample(int(0.001 * len(df)))
lat_lon_pairs = list(zip(list(sample_df.Start_Lat), list(sample_df.Start_Lng)))
```

Lets check the heat map to visualize the number of accidents that has occured in the united stated

```python
map = folium.Map()
HeatMap(lat_lon_pairs).add_to(map)
map
```
